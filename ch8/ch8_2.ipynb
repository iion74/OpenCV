```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
img1 = cv2.imread('/content/gdrive/MyDrive/Images/scaning.jpg')

RGVB_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGVB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
```

```python
kernel = np.ones((4,1), np.uint8)
dilateimg = cv2.dilate(gimg, kernel, iterations = 1)
cv2_imshow(dilateimg)
```

```python
kernel = np.ones((3,1), np.uint8)
erodeimg = cv2.erode(dilateimg, kernel, iterations = 1)
cv2_imshow(erodeimg)
```
